```julia
# Assignment problem(revised)
using JuMP
m=Model()
c1=[2000 2300; 1800 2100; 2400 2200;]
c2=[1100 850 1200 1000;1200 1000 1100 900;]
c3=[375*1000; 390*1000; 365*1000;]
@variable(m,x1[i=1:3,j=1:2] >= 1)
@variable(m,x2[i=1:2,j=1:4] >= 1)
@objective(m,Min,sum(c1.*x1[i=1:3,j=1:2]) + sum(c2.*x2[i=1:2,j=1:4]) +  sum(c3.*x1[i=1:3,j=1]) + sum(c3.*x1[i=1:3,j=2]))

#capacity constraint for plants
@constraint(m, [i=1],sum(x1[i,j] for j=1:2)<= 30 )
@constraint(m, [i=1],sum(x1[i,j] for j=1:2)>= 20)
@constraint(m, [i=2],sum(x1[i,j] for j=1:2)<= 40)
@constraint(m, [i=2],sum(x1[i,j] for j=1:2)>= 25)
@constraint(m, [i=3],sum(x1[i,j] for j=1:2)<= 30)
@constraint(m, [i=3],sum(x1[i,j] for j=1:2)>= 15)

#capacity constraint for ships
@constraint(m,  5 <= x1[1,1] <= 20)
@constraint(m,  10 <= x1[2,1] <= 30)
@constraint(m,  5 <= x1[3,1] <= 25)
@constraint(m,  5 <= x1[1,2] <= 20)
@constraint(m,  10 <= x1[2,2] <= 30)
@constraint(m,  5 <= x1[3,2] <= 15)
                     

@constraint(m, [j=1:1],sum(x1[i,j] for i=1:3) >= 20)
@constraint(m, [j=1:1],sum(x1[i,j] for i=1:3) <= 65)
@constraint(m, [j=2:2],sum(x1[i,j] for i=1:3) >= 30)
@constraint(m, [j=2:2],sum(x1[i,j] for i=1:3) <= 70)

                                        
#capacity constraint for trucks
@constraint(m, 3 <= x2[1,1] <= 7)
@constraint(m, 6 <= x2[1,2] <= 12)
@constraint(m, 10 <= x2[1,3] <= 18)
@constraint(m, 5 <= x2[1,4] <= 15)
@constraint(m, 4 <= x2[2,1] <= 6)
@constraint(m, 5 <= x2[2,2] <= 14)                                       
@constraint(m, 5 <= x2[2,3] <= 20)
@constraint(m, 10 <= x2[2,4] <= 20)


                                        
#demand constraint 
@constraint(m, [j=1],sum(x2[i,j] for i=1:2) == 10)
@constraint(m, [j=2],sum(x2[i,j] for i=1:2) == 20)
@constraint(m, [j=3],sum(x2[i,j] for i=1:2) == 30)
@constraint(m, [j=4],sum(x2[i,j] for i=1:2) == 25)
    
#Flow constraint
@constraint(m, sum(x1[i,1] for i=1:3) == sum(x2[1,j] for j=1:4))  
@constraint(m, sum(x1[i,2] for i=1:3) == sum(x2[2,j] for j=1:4))                                                         
                                                        
println(m)
                                                        
status=solve(m)


println("x1=",getvalue(x1[i=1:3,j=1:2]))
println("x2=",getvalue(x2[i=1:2,j=1:4]))
println("Objective value=",getobjectivevalue(m))
                                                        

```
